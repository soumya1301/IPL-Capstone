## Some Questions

### 1. Which team won the most matches?

```python
df['match_winner'].value_counts()
```

```python
match_wins=df['match_winner'].value_counts()
sns.barplot(y= match_wins.index, x=match_wins.values, palette='viridis')
plt.title("Most Matches Won by Team")
```

### 2. Toss Decision Trends

```python
sns.countplot(x=df['toss_decision'],palette='rainbow')
plt.title("Decision Trends")
```

### 3. Toss Winner VS Match Winner

```python
df[df['toss_winner']==df['match_winner']]
```

```python
count= df[df['toss_winner']==df['match_winner']]['match_id'].count()
percentage=(count*100)/df.shape[0]
percentage
```

### 4. How do teams win? (Runs VS Wickets)

```python
sns.countplot(x=df['won_by'],palette='rocket')
plt.title("Won by")
```

## Key Player Performance

### 1. Most Player of the match award

```python
counts=df['player_of_the_match'].value_counts().head(10)
counts
```

```python
sns.barplot(x= counts.values, y=counts.index, palette='mako')
plt.title("Top 10 players with man of the match")
```

### 2. Two Top Scores

```python
high=df.groupby('top_scorer')
high
```

```python
high=df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind='barh')
plt.title("Top two scorers")
```

### 3. Ten Best Bowling Figures

```python
df.head()
```

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x :x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
df.head()
# made a new column by the name highest_wickets and changed its type from string to integer
```

```python
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers.plot(kind = 'barh')
```
